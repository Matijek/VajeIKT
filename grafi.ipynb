```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import FileLink

# Nastavitve grafov
plt.rcParams['font.size'] = 12
plt.rcParams['axes.labelsize'] = 10
plt.rcParams['axes.titlesize'] = 12

# Podatki
x = np.linspace(0, 10, 500)  # Prilagodite območje in korak x, kot želite
f1 = np.cos(x)
f2 = f1 ** 2
g1 = np.cos(x) * np.exp(-0.1 * x)
g2 = g1 ** 2

# Narišemo grafe f_1(x) in f_2(x) na istem grafu
plt.figure(figsize=(15, 6.5))

plt.plot(x, f1, label='$f_1(x) = \\cos(x)$')
plt.plot(x, f2, label='$f_2(x) = (f_1(x))^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafa $f_1(x)$ in $f_2(x)$')
plt.legend()

# Izvoz slike v jpg format
plt.savefig('f_graphs.jpg', dpi=300, bbox_inches='tight')
FileLink('f_graphs.jpg')

# Narišemo grafe g_1(x) in g_2(x) na istem grafu
plt.figure(figsize=(15, 6.5))

plt.plot(x, g1, label='$g_1(x) = \\cos(x) \\cdot e^{-0.1x}$')
plt.plot(x, g2, label='$g_2(x) = (g_1(x))^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafa $g_1(x)$ in $g_2(x)$')
plt.legend()

# Izvoz slike v jpg format
plt.savefig('g_graphs.jpg', dpi=300, bbox_inches='tight')

# Prikažemo grafe
plt.show()

# Generate download links for the files

FileLink('g_graphs.jpg')
```
